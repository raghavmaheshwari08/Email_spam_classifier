#Importing the Libraries

```python
import numpy as np
import pandas as pd
```

#Loading the Dataset

```python
dataset = pd.read_csv('spam.csv', encoding='latin-1')
```

#Data Preparation and Preprocessing

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
label_map = {
    'spam' : 1,
    'ham' : 0
}
```

```python
dataset['Category'] = dataset['Category'].replace(label_map).infer_objects(copy=False)
```

```python
dataset.head()
```

```python
# Creating a function to Remove special characters from the data
def rem_special_chars(text):
    new_text = ""
    for i in text:
        if i.isalnum() or i == " ":
            new_text += i
    return new_text.strip()
```

```python
# Removing special characters from the data
dataset['Message'] = dataset['Message'].apply(rem_special_chars)
```

```python
dataset.head()
```

```python
# Convert all to small alphabet letters
dataset['Message'] = dataset['Message'].apply(lambda x:x.lower())
```

```python
dataset.head()
```

```python
X = dataset['Message']
y = dataset['Category']
```

#Applying CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english',max_features=10000)
```

```python
X = cv.fit_transform(dataset['Message']).toarray()
```

```python
X
```

```python
y = dataset['Category']
```

```python
y.values
```

#Splitting the Dataset to Training an Test Set

```python
# Splitting the data into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

#Applying Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

#Computing the Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```
